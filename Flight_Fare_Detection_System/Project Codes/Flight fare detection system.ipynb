```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
train_data=pd.read_excel(r"Data_Train.xlsx")
```

```python
train_data.head()
```

```python
train_data.info() #here price is our dependent feature(int64 type) while all other paramaters are our independent features(string type)
```

```python
train_data["Duration"].value_counts() #unique value counts;
```

```python
train_data.shape
```

```python
train_data.dropna(inplace = True)
```

```python
train_data.isnull().sum() #to check the count of null values in every column (we get it as 0 so there are no null values)
```

```python
train_data["Journey_day"]=pd.to_datetime(train_data.Date_of_Journey,format="%d/%m/%Y").dt.day 
#Adding new column for journeey day in the datasheet
```

```python
train_data["Journey_month"] = pd.to_datetime(train_data.Date_of_Journey, format = "%d/%m/%Y").dt.month 
#Adding new column for journey month in the datasheet
```

```python
train_data.head()
```

```python
train_data.drop(["Date_of_Journey"], axis = 1, inplace = True)
#dropping date of journey since it is of no use
```

```python
#Similarly doing same process with Dep_Time column and extrating them into Dep_hour and Dep_min;
train_data["Dep_hour"] = pd.to_datetime(train_data["Dep_Time"]).dt.hour
train_data["Dep_min"] = pd.to_datetime(train_data["Dep_Time"]).dt.minute

# Now dropping Dep_Time as it is of no use for our data analysis; 
train_data.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
train_data.head()
```

```python
# Similarly doing same operation on Arrival_Time
train_data["Arrival_hour"] = pd.to_datetime(train_data.Arrival_Time).dt.hour
train_data["Arrival_min"] = pd.to_datetime(train_data.Arrival_Time).dt.minute
# Now we can drop Arrival_Time as it is of no use
train_data.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
train_data.head() #here we get two more rows with Arrival_hour and Arrival_min
```

```python
# Converting Duration column into list
duration = list(train_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2: # The entities which do no have min portion adding 0m in then and vice-versa;
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = [] # List for duration hours
duration_mins = [] #List for duration mins
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
```

```python
#Adding thesse columns into dataset;
train_data["Duration_hours"] = duration_hours
train_data["Duration_mins"] = duration_mins
```

```python
train_data.drop(["Duration"], axis = 1, inplace = True)
#now duration is of no use so dropping it off!!
```

```python
train_data.head() # we can see two extra columns has been added
```

```python
train_data["Airline"].value_counts()
```

```python
# Categorical plot to observe the relation between the Airline and their respective price ;
sns.catplot(y = "Price", x = "Airline", data = train_data.sort_values("Price", ascending = False), kind="boxen", height = 6, aspect = 3)
plt.show()

# We can get one conclusion that the price of jetairway business is high as compared to others;
```

```python
# Now we came to know that the data cant be compared with arithmetic operations so we will perform OneHotEncoding;
#Wherever the airline values is present we will get 1 else we will get 0 in the columns 
#This ensures easy proessing of data;
Airline = train_data[["Airline"]]
Airline = pd.get_dummies(Airline, drop_first= True)
Airline.head()
```

```python
# The operations we did for Airline column same operations we will do for Source column
train_data["Source"].value_counts()
```

```python
sns.catplot(y = "Price", x = "Source", data = train_data.sort_values("Price", ascending = False), kind="boxen", height = 4, aspect = 3)
plt.show()
#Source vs Price
```

```python
#Wherever the Source values is present we will get 1 else we will get 0 in the columns 
#This ensures easy proessing of data;

Source = train_data[["Source"]]
Source = pd.get_dummies(Source, drop_first= True)
Source.head()
```

```python
train_data["Destination"].value_counts()
```

```python
# Similarly performing same operations on Destination column

Destination = train_data[["Destination"]]

Destination = pd.get_dummies(Destination, drop_first = True)

Destination.head()
```

```python
#Now the route column and Total_Stops are proiding the same data ie(count of stops)
#so there is no need of it so we will drop the Route entity!
train_data.drop(["Route"], axis = 1, inplace = True)
```

```python
#Similarly the Additional_info column is also of no use to use so we will drop this column too;
train_data.drop(["Additional_Info"], axis = 1, inplace = True)
```

```python
#This gives the no of stops in our data;
train_data["Total_Stops"].value_counts()
```

```python
#As we can see there are atmost 4 stops in the data so we can perform label encoding on this data
#and proceed with numerical values for easy comparison
train_data.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
train_data.head() # Now we can see most of the data is converted into numerical format and can be used for analysis;
```

```python
#We have converted all the columnns into numerical values so there is no need of (Airline, Source, Destination) so we
#will concatinate them and delete them at once;
data_train = pd.concat([train_data, Airline, Source, Destination], axis = 1)
```

```python
data_train.head()
```

```python
data_train.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True)
```

```python
data_train.head()
#Soo finaly we got one clean dataset with all the numerical values for analyzing the fare;
```

```python
# Performing same operation on test_data set that we have performed on train_data set;
# We need to perform the operations diffrently so that there is no data leakage;
test_data = pd.read_excel(r"Data_Train.xlsx")
```

```python
test_data.head()
```

```python
  
#Here basically we are performing the same task that we have performed for train_data set;
print("Test data Info")
print("-"*75)
print(test_data.info())

print()
print()

print("Null values :")
print("-"*75)
test_data.dropna(inplace = True)
print(test_data.isnull().sum())

# Date_of_Journey
test_data["Journey_day"] = pd.to_datetime(test_data.Date_of_Journey, format="%d/%m/%Y").dt.day
test_data["Journey_month"] = pd.to_datetime(test_data["Date_of_Journey"], format = "%d/%m/%Y").dt.month
test_data.drop(["Date_of_Journey"], axis = 1, inplace = True)

# Dep_Time
test_data["Dep_hour"] = pd.to_datetime(test_data["Dep_Time"]).dt.hour
test_data["Dep_min"] = pd.to_datetime(test_data["Dep_Time"]).dt.minute
test_data.drop(["Dep_Time"], axis = 1, inplace = True)

# Arrival_Time
test_data["Arrival_hour"] = pd.to_datetime(test_data.Arrival_Time).dt.hour
test_data["Arrival_min"] = pd.to_datetime(test_data.Arrival_Time).dt.minute
test_data.drop(["Arrival_Time"], axis = 1, inplace = True)

# Duration
duration = list(test_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration

# Adding Duration column to test set
test_data["Duration_hours"] = duration_hours
test_data["Duration_mins"] = duration_mins
test_data.drop(["Duration"], axis = 1, inplace = True)


# Categorical data

print("Airline")
print("-"*75)
print(test_data["Airline"].value_counts())
Airline = pd.get_dummies(test_data["Airline"], drop_first= True)

print()

print("Source")
print("-"*75)
print(test_data["Source"].value_counts())
Source = pd.get_dummies(test_data["Source"], drop_first= True)

print()

print("Destination")
print("-"*75)
print(test_data["Destination"].value_counts())
Destination = pd.get_dummies(test_data["Destination"], drop_first = True)

# Additional_Info is of no use for us;
# Route and Total_Stops are related to each other as they provide only one data ie no of stops so we are
#dropping off the column

test_data.drop(["Route", "Additional_Info"], axis = 1, inplace = True)

# Replacing Total_Stops (Label encoding)
test_data.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)

# Concatenate dataframe --> test_data + Airline + Source + Destination (Concatination for deletion process)
data_test = pd.concat([test_data, Airline, Source, Destination], axis = 1)
data_test.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True)

print()
print()

print("Shape of test data : ", data_test.shape)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = reg_rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
prediction = rf_random.predict(X_test)
```

```python
plt.figure(figsize = (8,8))
sns.distplot(y_test-prediction)
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.scatter(y_test, prediction, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```
